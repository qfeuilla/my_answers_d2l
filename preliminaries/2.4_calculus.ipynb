** Question 3 : What is the gradient of the function $ f(x)=\left \| x \right \|_2 $?**

$ \left \| x \right \|_2 = \sqrt{x^\top x} $

let $ u = x^\top x$

then using the chain rule :

$ \nabla x f(x) = \frac{\partial u^\frac{1}{2}}{\partial u} * \frac{\partial u}{\partial x}  $

$ \nabla x f(x) = (\frac{1}{2} * u^{-\frac{1}{2}} ) * 2x $

$ \nabla x f(x) = \frac{2x}{2 * \sqrt{x^\top x} } $

$ \nabla x f(x) = \frac{x}{\left \| x \right \|_2} $
