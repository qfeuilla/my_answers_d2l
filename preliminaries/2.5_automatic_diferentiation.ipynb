```python
import torch
import matplotlib.pyplot as plt

x = torch.arange(4.0, requires_grad=True)
y = 2 * torch.dot(x, x)
y.backward()
x.grad
```

```python
def f(x):
    return torch.sin(x)
```

```python
x = torch.arange(0, 10, 0.2, requires_grad=True)
_fx = f(x)
for i in range(_fx.shape[0]):
    _fx[i].backward(retain_graph = True)
plt.plot(_fx.detach())
plt.plot(x.grad)
plt.show()
```
